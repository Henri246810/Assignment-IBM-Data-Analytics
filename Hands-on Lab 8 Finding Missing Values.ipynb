### 2. Explore the Dataset
##### Task 1: Display basic information and summary statistics of the dataset.


```python
## Write your code here
import pandas as pd

file_path = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"
df = pd.read_csv(file_path)

# Display basic information about the dataset
print("Basic Information:")
print(df.info())

# Display the number of rows and columns
print("\nDataset Shape:", df.shape)

# Display column names
print("\nColumn Names:")
print(df.columns.tolist())

# Display summary statistics for numeric columns
print("\nSummary Statistics (Numeric Columns):")
print(df.describe())

# Display summary statistics for non-numeric columns
print("\nSummary Statistics (Non-Numeric Columns):")
print(df.describe(include=['object']))
```

### 3. Finding Missing Values
##### Task 2: Identify missing values for all columns.


```python
## Write your code here
# Count missing values for each column
missing_counts = df.isna().sum()

# Display only columns with missing values, sorted from most to least
missing_counts = missing_counts[missing_counts > 0].sort_values(ascending=False)
print("Columns with missing values and their counts:")
print(missing_counts)
```

##### Task 3: Visualize missing values using a heatmap (Using seaborn library).



```python
## Write your code here
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))
sns.heatmap(df.isna(), cbar=False, yticklabels=False, cmap='viridis')
plt.title("Heatmap of Missing Values")
plt.show()
```

##### Task 4: Count the number of missing rows for a specific column (e.g., `Employment`).


```python
## Write your code here
missing_employment = df['Employment'].isna().sum()
print("Number of missing rows in 'Employment':", missing_employment)
```

### 4. Imputing Missing Values
##### Task 5: Identify the most frequent (majority) value in a specific column (e.g., `Employment`).


```python
## Write your code here
# Find the most frequent value in the 'Employment' column
most_frequent = df['Employment'].mode()[0]
print("Most frequent value in 'Employment':", most_frequent)
```

##### Task 6: Impute missing values in the `Employment` column with the most frequent value.



```python
# Find the most frequent value
most_frequent = df['Employment'].mode()[0]

# Impute missing values by assigning back to the column
df['Employment'] = df['Employment'].fillna(most_frequent)

# Verify no missing values remain
missing_employment = df['Employment'].isna().sum()
print("Number of missing rows in 'Employment' after imputation:", missing_employment)
```

### 5. Visualizing Imputed Data
##### Task 7: Visualize the distribution of a column after imputation (e.g., `Employment`).


```python
## Write your code here
import matplotlib.pyplot as plt

# Count the occurrences of each category in 'Employment'
employment_counts = df['Employment'].value_counts()

# Plot the distribution as a bar chart
plt.figure(figsize=(8,5))
employment_counts.plot(kind='bar', color='skyblue')
plt.title('Distribution of Employment After Imputation')
plt.xlabel('Employment Type')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```
