### Step 2: Load the Dataset


```python
# Load the dataset from the given URL
file_url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"
df = pd.read_csv(file_url)

# Display the first few rows to understand the structure of the dataset
df.head()
```

<h3>Step 3: Analyze and Visualize Compensation Distribution</h3>


**Task**: Plot the distribution and histogram for `ConvertedCompYearly` to examine the spread of yearly compensation among respondents.


```python
## Write your code here
import matplotlib.pyplot as plt

# Ensure ConvertedCompYearly is numeric
df['ConvertedCompYearly'] = pd.to_numeric(df['ConvertedCompYearly'], errors='coerce')

# Drop missing values for plotting
comp_data = df['ConvertedCompYearly'].dropna()

# Summary statistics
print(comp_data.describe())

# Histogram
plt.figure(figsize=(10,6))
plt.hist(comp_data, bins=50, color='skyblue', edgecolor='black')
plt.title('Distribution of ConvertedCompYearly')
plt.xlabel('Annual Compensation (USD)')
plt.ylabel('Number of Respondents')
plt.xlim(0, comp_data.quantile(0.95))  # Zoom in to 95th percentile to reduce skew
plt.show()
```

<h3>Step 4: Calculate Median Compensation for Full-Time Employees</h3>


**Task**: Filter the data to calculate the median compensation for respondents whose employment status is "Employed, full-time."


```python
## Write your code here
# Filter for full-time employees
full_time = df[df['Employment'] == 'Employed, full-time']

# Ensure compensation is numeric
full_time['ConvertedCompYearly'] = pd.to_numeric(full_time['ConvertedCompYearly'], errors='coerce')

# Calculate median compensation
median_comp = full_time['ConvertedCompYearly'].median()
print(f"Median annual compensation for full-time employees: ${median_comp:,.2f}")
```

<h3>Step 5: Analyzing Compensation Range and Distribution by Country</h3>


Explore the range of compensation in the ConvertedCompYearly column by analyzing differences across countries. Use box plots to compare the compensation distributions for each country to identify variations and anomalies within each region, providing insights into global compensation trends.



```python
## Write your code here
import matplotlib.pyplot as plt
import seaborn as sns

# Ensure numeric compensation
df['ConvertedCompYearly'] = pd.to_numeric(df['ConvertedCompYearly'], errors='coerce')

# Select countries with enough data to avoid clutter
top_countries = df['Country'].value_counts().head(10).index
df_top_countries = df[df['Country'].isin(top_countries)]

# Plot boxplot of compensation by country
plt.figure(figsize=(12,6))
sns.boxplot(x='Country', y='ConvertedCompYearly', data=df_top_countries)
plt.title('Compensation Distribution by Country (Top 10 Countries)')
plt.ylabel('Annual Compensation (USD)')
plt.xticks(rotation=45)
plt.ylim(0, df_top_countries['ConvertedCompYearly'].quantile(0.95))  # Limit to 95th percentile
plt.show()
```

<h3>Step 6: Removing Outliers from the Dataset</h3>


**Task**: Create a new DataFrame by removing outliers from the `ConvertedCompYearly` column to get a refined dataset for correlation analysis.


```python
## Write your code here
# Ensure numeric compensation
df['ConvertedCompYearly'] = pd.to_numeric(df['ConvertedCompYearly'], errors='coerce')

# Drop rows with missing compensation
comp_data = df.dropna(subset=['ConvertedCompYearly'])

# Calculate IQR
Q1 = comp_data['ConvertedCompYearly'].quantile(0.25)
Q3 = comp_data['ConvertedCompYearly'].quantile(0.75)
IQR = Q3 - Q1

# Define bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter dataset to exclude outliers
df_clean = comp_data[(comp_data['ConvertedCompYearly'] >= lower_bound) & 
                     (comp_data['ConvertedCompYearly'] <= upper_bound)]

# Verify the size reduction
print(f"Original dataset: {comp_data.shape[0]} rows")
print(f"Dataset after removing outliers: {df_clean.shape[0]} rows")
```
