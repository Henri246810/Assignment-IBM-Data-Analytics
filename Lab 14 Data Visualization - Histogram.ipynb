## Hands-on Lab: Visualizing Data with Histograms


### 1. Visualizing the distribution of data (Histograms)


**1.1 Histogram of `CompTotal` (Total Compensation)**


Objective: Plot a histogram of `CompTotal` to visualize the distribution of respondents' total compensation.


```python
import matplotlib.pyplot as plt

# Step 1: Check if CompTotal exists
if 'CompTotal' in df.columns:
    
    # Step 2: Drop missing values
    comp_data = df['CompTotal'].dropna()
    
    # Step 3: Optional - remove extreme outliers for visualization
    # For example, keep only values within 1st to 99th percentile
    lower = comp_data.quantile(0.01)
    upper = comp_data.quantile(0.99)
    comp_data_filtered = comp_data[(comp_data >= lower) & (comp_data <= upper)]
    
    # Step 4: Plot histogram
    plt.figure(figsize=(10,6))
    plt.hist(comp_data_filtered, bins=30, color='skyblue', edgecolor='black')
    plt.title('Distribution of Total Compensation (CompTotal)')
    plt.xlabel('CompTotal (USD)')
    plt.ylabel('Number of Respondents')
    plt.grid(axis='y', alpha=0.75)
    plt.show()
    
else:
    print("Column 'CompTotal' does not exist in the dataset.")
```

**1.2 Histogram of YearsCodePro (Years of Professional Coding Experience)**


Objective: Plot a histogram of `YearsCodePro` to analyze the distribution of coding experience among respondents.


```python
import matplotlib.pyplot as plt

# Step 1: Check if YearsCodePro exists
if 'YearsCodePro' in df.columns:
    
    # Step 2: Drop missing values
    years_data = df['YearsCodePro'].dropna()
    
    # Step 3: Convert to numeric if needed (some entries might be strings)
    years_data = pd.to_numeric(years_data, errors='coerce').dropna()
    
    # Step 4: Optional - remove extreme outliers (1st–99th percentile)
    lower = years_data.quantile(0.01)
    upper = years_data.quantile(0.99)
    years_data_filtered = years_data[(years_data >= lower) & (years_data <= upper)]
    
    # Step 5: Plot histogram
    plt.figure(figsize=(10,6))
    plt.hist(years_data_filtered, bins=20, color='lightgreen', edgecolor='black')
    plt.title('Distribution of Professional Coding Experience (YearsCodePro)')
    plt.xlabel('Years of Professional Coding')
    plt.ylabel('Number of Respondents')
    plt.grid(axis='y', alpha=0.75)
    plt.show()
    
else:
    print("Column 'YearsCodePro' does not exist in the dataset.")
```

### 2. Visualizing Relationships in Data


**2.1 Histogram Comparison of `CompTotal` by `Age` Group**


Objective: Use histograms to compare the distribution of CompTotal across different Age groups.


```python
import matplotlib.pyplot as plt

# Step 1: Ensure the required columns exist
if 'CompTotal' in df.columns and 'Age' in df.columns:
    
    # Step 2: Drop rows with missing values in CompTotal or Age
    comp_age_df = df[['CompTotal', 'Age']].dropna()
    
    # Step 3: Optional - convert CompTotal to numeric and remove extreme outliers
    comp_age_df['CompTotal'] = pd.to_numeric(comp_age_df['CompTotal'], errors='coerce')
    comp_age_df = comp_age_df.dropna()
    lower = comp_age_df['CompTotal'].quantile(0.01)
    upper = comp_age_df['CompTotal'].quantile(0.99)
    comp_age_df = comp_age_df[(comp_age_df['CompTotal'] >= lower) & (comp_age_df['CompTotal'] <= upper)]
    
    # Step 4: Plot histograms for each Age group
    age_groups = comp_age_df['Age'].unique()
    plt.figure(figsize=(12,6))
    
    for age in age_groups:
        data = comp_age_df[comp_age_df['Age'] == age]['CompTotal']
        plt.hist(data, bins=30, alpha=0.5, label=age)
    
    plt.title('Comparison of CompTotal by Age Group')
    plt.xlabel('CompTotal (USD)')
    plt.ylabel('Number of Respondents')
    plt.legend(title='Age Group', bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.grid(axis='y', alpha=0.75)
    plt.tight_layout()
    plt.show()
    
else:
    print("Required columns 'CompTotal' or 'Age' are missing in the dataset.")
```

**2.2 Histogram of TimeSearching for Different Age Groups**


Objective: Use histograms to explore the distribution of `TimeSearching` (time spent searching for information) for respondents across different age groups.


```python
import matplotlib.pyplot as plt

# Step 1: Ensure required columns exist
if 'TimeSearching' in df.columns and 'Age' in df.columns:
    
    # Step 2: Drop rows with missing values
    time_age_df = df[['TimeSearching', 'Age']].dropna()
    
    # Step 3: Convert TimeSearching to numeric and drop invalid rows
    time_age_df['TimeSearching'] = pd.to_numeric(time_age_df['TimeSearching'], errors='coerce')
    time_age_df = time_age_df.dropna()
    
    # Step 4: Plot histograms for each Age group
    age_groups = time_age_df['Age'].unique()
    plt.figure(figsize=(12,6))
    
    plotted_any = False  # flag to track if any histogram is actually plotted
    for age in age_groups:
        data = time_age_df[time_age_df['Age'] == age]['TimeSearching']
        if not data.empty:
            plt.hist(data, bins=20, alpha=0.5, label=age)
            plotted_any = True
    
    if plotted_any:
        plt.title('Distribution of Time Spent Searching by Age Group')
        plt.xlabel('Time Spent Searching (hours per week)')
        plt.ylabel('Number of Respondents')
        plt.legend(title='Age Group', bbox_to_anchor=(1.05, 1), loc='upper left')
        plt.grid(axis='y', alpha=0.75)
        plt.tight_layout()
        plt.show()
    else:
        print("No valid data to plot after cleaning.")
    
else:
    print("Required columns 'TimeSearching' or 'Age' are missing in the dataset.")
```
